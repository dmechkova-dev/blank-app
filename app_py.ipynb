```python
import streamlit as st

st.set_page_config(page_title="Python Р·Р° 6. РәР»Р°СҒ", page_icon="рҹҗҚ")

st.title("рҹҡҖ РҹСҖРёРәР»СҺСҮРөРҪРёРө РІ СҒРІРөСӮР° РҪР° Python!")
st.write("Р”РҫРұСҖРө РҙРҫСҲР»Рё, РјР»Р°РҙРё РҝСҖРҫРіСҖР°РјРёСҒСӮРё! РқРөРәР° РҪР°СғСҮРёРј РҫСҒРҪРҫРІРёСӮРө Р·Р°РөРҙРҪРҫ.")

# РңРөРҪСҺ РІСҒСӮСҖР°РҪРё
choice = st.sidebar.selectbox("РҳР·РұРөСҖРё СӮРөРјР°:", ["РўРёРҝРҫРІРө РҙР°РҪРҪРё", "РңР°СӮРөРјР°СӮРёРәР°", "РҰРёРәР»Рё (Loops)"])

if choice == "РўРёРҝРҫРІРө РҙР°РҪРҪРё":
    st.header("рҹ“Ұ РўРёРҝРҫРІРө РҝСҖРҫРјРөРҪР»РёРІРё")
    st.write("РҹСҖРҫРјРөРҪР»РёРІРёСӮРө СҒР° РәР°СӮРҫ РәСғСӮРёРё, РІ РәРҫРёСӮРҫ РҝР°Р·РёРј РёРҪС„РҫСҖРјР°СҶРёСҸ.")

    name = st.text_input("РқР°РҝРёСҲРё РёРјРөСӮРҫ СҒРё:", "Р РҫРұРҫСӮ")
    age = st.number_input("РқР°РҝРёСҲРё РіРҫРҙРёРҪРёСӮРө СҒРё:", value=12)

    if st.button("РҹРҫРәР°Р¶Рё СӮРёРҝР°!"):
        st.success(f"РўРІРҫРөСӮРҫ РёРјРө '{name}' Рө СӮРёРҝ **String** (СӮРөРәСҒСӮ)")
        st.info(f"РўРІРҫРёСӮРө РіРҫРҙРёРҪРё {age} СҒР° СӮРёРҝ **Integer** (СҶСҸР»Рҫ СҮРёСҒР»Рҫ)")

elif choice == "РңР°СӮРөРјР°СӮРёРәР°":
    st.header("рҹ§® РҗСҖРёСӮРјРөСӮРёСҮРҪРё РҙРөР№СҒСӮРІРёСҸ")
    num1 = st.number_input("Р§РёСҒР»Рҫ Рҗ:", value=10)
    num2 = st.number_input("Р§РёСҒР»Рҫ B:", value=5)

    st.write(f"**{num1} + {num2} = {num1 + num2}**")
    st.write(f"**{num1} - {num2} = {num1 - num2}**")
    st.write(f"**{num1} * {num2} = {num1 * num2}**")
    st.write(f"**{num1} / {num2} = {num1 / num2}**")

elif choice == "РҰРёРәР»Рё (Loops)":
    st.header("рҹ”„ РҰРёРәР»Рё")
    st.write("РҰРёРәСҠР»СҠСӮ `for` РҝРҫРІСӮР°СҖСҸ РҪРөСүРҫ РјРҪРҫРіРҫ РҝСҠСӮРё.")

    times = st.slider("РҡРҫР»РәРҫ РҝСҠСӮРё РҙР° РҝРҫРІСӮРҫСҖСҸ?", 1, 10, 5)
    word = st.text_input("РҡР°РәРІРҫ РҙР° РәР°Р¶Р°?", "Python Рө СҒСғРҝРөСҖ!")

    if st.button("РЎСӮР°СҖСӮ!"):
        for i in range(times):
            st.write(f"{i+1}. {word}")
```
